
# Unit 4: The Hessian Matrix (Peaks, Valleys, and Saddles)

In 2D calculus, finding a peak or a valley was a two-step process:

1. Find where the first derivative is zero (a flat spot).
2. Check the **second derivative** (concavity). If it curves up, it's a valley. If it curves down, it's a peak.

In 3D, things are more complicated. If you are standing on a mountain pass (a saddle point), the terrain might curve *up* in the X-direction, but curve *down* in the Y-direction.

To figure out the true shape of a 3D surface at a flat spot, we have to calculate all the possible second derivatives and organize them into a 2x2 grid called the **Hessian Matrix**.

Let's load our libraries and build our second-derivative tools!


```python
import numpy as np
import matplotlib.pyplot as plt

# Here are the first derivative tools from previous notebooks
def partial_x(f, x, y, h=1e-5):
    return (f(x + h, y) - f(x, y)) / h

def partial_y(f, x, y, h=1e-5):
    return (f(x, y + h) - f(x, y)) / h
```
