# Unit 5: Line Integrals and Work

In introductory physics, you learn that **Work = Force $\times$ Distance**. If you push a box 10 meters with a force of 5 Newtons, you did 50 Joules of work.

But what if the path isn't a straight line? What if you are flying a drone through a twisting wind storm, and the wind is constantly changing direction and speed?

To calculate the total work, we have to chop our curved path into microscopic straight lines ($d\vec{r}$), find the wind vector at each specific point ($\vec{F}$), and use the **Dot Product** to see how much of the wind is helping us or fighting us. Then, we integrate (sum) them all up!

This is called a **Line Integral**:


$$W = \int_C \vec{F} \cdot d\vec{r}$$
