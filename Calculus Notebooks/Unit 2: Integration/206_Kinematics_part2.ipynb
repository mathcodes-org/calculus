
# Unit 2, Application 2: Kinematics Continued (Moving Backward)

In Unit 1, we learned that if we have a function for position $s(t)$, we can take the derivative to find velocity $v(t)$, and take the derivative again to find acceleration $a(t)$.

But in the real world, we rarely start with position. If you are building a video game or a rocket guidance system, you usually only know the forces acting on the object (like gravity), which means you only know the **acceleration**.

To find out where the object actually goes, we have to work backward using integration:

* $\int a(t) dt = v(t) + C$
* $\int v(t) dt = s(t) + C$

That $+ C$ is the **constant of integration**. In physics, it represents our **initial conditions**: the initial velocity $v(0)$ and the initial position $s(0)$.

Let's load our libraries and numerical integration tool.


```python
import numpy as np
import matplotlib.pyplot as plt

def numerical_integration(f_data, dx):
    """
    Takes an array of y-values and the step size dx.
    Returns the accumulated area array.
    """
    # Calculate the area of each microscopic rectangle
    tiny_areas = f_data * dx
    # Return the cumulative sum
    return np.cumsum(tiny_areas)
```
