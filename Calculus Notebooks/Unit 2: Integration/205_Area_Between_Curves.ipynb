# Unit 2, Application: Area Between Curves

We have reached the capstone of Unit 2. Up to this point, we have been calculating the area between a function and the x-axis. But what if we want to find the area trapped between two completely different functions?

Mathematically, the area between an upper curve $f(x)$ and a lower curve $g(x)$ from $x=a$ to $x=b$ is found by subtracting them and integrating:


$$Area = \int_a^b |f(x) - g(x)| dx$$

But to do this, we need to know exactly where the curves intersect to find our bounds $a$ and $b$. Today, you will write a single master script that automates this entire process.

Let's load our libraries and bring in the two most powerful numerical tools we built in previous notebooks.


```python
import numpy as np
import matplotlib.pyplot as plt

# Tool 1: Finds where an array crosses zero
def find_roots(x_array, y_array):
    sign_changes = y_array[:-1] * y_array[1:] < 0
    return x_array[:-1][sign_changes]

# Tool 2: Calculates area
def trapezoidal_rule(f_data, dx):
    left_heights = f_data[:-1]
    right_heights = f_data[1:]
    areas = ((left_heights + right_heights) / 2) * dx
    return np.sum(areas)
```
