# Unit 2: Statistics and Probability

Welcome to continuous statistics. If you want to know the probability of an event happening in a continuous system, you cannot use basic fractions. You have to use calculus.

In statistics, we use a special function called a **Probability Density Function (PDF)**, usually written as $f(x)$. The PDF shows the relative likelihood of different outcomes.

Because it represents probability, a PDF has one unbreakable rule: **The total area under the entire curve must equal exactly 1 (which represents 100% probability).** If you want to find the probability of an outcome falling between $a$ and $b$, you simply integrate the PDF from $a$ to $b$.

Let's load our libraries and our trusty integration tool.

```python
import numpy as np
import matplotlib.pyplot as plt

def trapezoidal_rule(f_data, dx):
    left_heights = f_data[:-1]
    right_heights = f_data[1:]
    areas = ((left_heights + right_heights) / 2) * dx
    return np.sum(areas)
```

## Challenge 3: Calculating Expected Value (The Mean)

If you have a dataset of test scores, you find the average (mean) by adding them up and dividing by the total count.

But if you only have a continuous probability curve, how do you find the mean? In calculus, the mean is called the **Expected Value**, written as $E[X]$ or $\mu$ (mu). You calculate it by multiplying every possible $x$-value by its probability $f(x)$, and summing them all up.

Mathematically, you just stick an $x$ inside the integral:


$$E[X] = \int_{-\infty}^{\infty} x \cdot f(x) dx$$

## Challenge 5: The General Normal Distribution

Real-world data is rarely centered exactly at zero. To handle real data, we have to modify our PDF formula to include the actual mean ($\mu$) and the actual standard deviation ($\sigma$).

The general formula for any normal distribution is:


$$f(x) = \frac{1}{\sigma\sqrt{2\pi}} e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$

**The Problem:**
