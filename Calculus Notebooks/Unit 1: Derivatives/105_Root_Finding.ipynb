# Unit 1: Root Finding (Newton's Method)

Finding where a function crosses the x-axis (where $f(x) = 0$) is one of the most common problems in mathematics. These points are called **roots**.

In algebra, you learned to factor polynomials or use the quadratic formula to find roots. But what happens if you have a function like $f(x) = \cos(x) - x$? You cannot isolate $x$ algebraically.

Instead, we use calculus. **Newton's Method** is an algorithm that uses tangent lines to computationally "slide" down a curve until it hits the x-axis.

Let's set up our workspace and bring in the `numerical_derivative` function we built earlier.


```python
import numpy as np
import matplotlib.pyplot as plt

# We redefine our derivative tool here so this notebook is self-contained
def numerical_derivative(f, x, h=1e-5):
    return (f(x + h) - f(x)) / h
```

## The Mathematics of Newton's Method

If we make a random guess $x_0$, it probably won't be the root. But if we draw a tangent line at $x_0$ and follow that straight line down to the x-axis, it will land closer to the actual root. We call this new point $x_1$.

If we repeat this process, $x_2$, $x_3$, and $x_4$ will rapidly converge exactly on the root.

The formula for Newton's Method is:


$$x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}$$
