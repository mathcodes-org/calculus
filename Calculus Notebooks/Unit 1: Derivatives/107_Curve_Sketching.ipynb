## Challenge 2: Finding Roots via Sign Changes

To find critical points and inflection points, we need to know where $f'(x) = 0$ and $f''(x) = 0$.

While we *could* use Newton's Method from Notebook 06, there is a much faster way when we are dealing with massive `numpy` arrays of data. If an array of $y$-values crosses the x-axis, the numbers must change from positive to negative, or from negative to positive. If we multiply adjacent numbers together, a sign change will always result in a negative number!

**Your Task:**

1. Look at the function `find_roots(x_array, y_array)` provided below. Read the comments to understand how it uses array slicing to efficiently find where the data crosses zero.
2. Define the polynomial: $P(x) = \frac{1}{4}x^4 - x^3 - 2x^2 + 3$.
3. Create a domain `x_vals` from $-3$ to $5$ with 1000 points.
4. Calculate the first derivative array (`y_prime`) and second derivative array (`y_double_prime`) using your tools.
5. Use the `find_roots` function to print the approximate $x$-values of the critical points and inflection points.


```python
def find_roots(x_array, y_array):
    """
    Finds the x-values where the y-array crosses zero.
    """
    # Multiply adjacent elements. If the sign changed, the product is negative!
    sign_changes = y_array[:-1] * y_array[1:] < 0
    
    # Return the x-values where those sign changes happened
    return x_array[:-1][sign_changes]

# 1. Define P(x)
# 2. Create x_vals
# 3. Calculate y_prime and y_double_prime arrays
# 4. Find and print the critical points and inflection points!
```
