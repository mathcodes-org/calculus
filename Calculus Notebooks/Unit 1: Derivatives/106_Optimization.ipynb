# Unit 1: Optimization 

In the previous notebook, we used Newton's Method to find where a function equals zero. But what if we want to find the lowest possible point (a minimum) or the highest possible point (a maximum) of a function?

Finding these points is called **optimization**, and it is the exact calculus concept that powers modern artificial intelligence and machine learning.

Let's set up our workspace and bring in our numerical derivative tool.


```python

import numpy as np
import matplotlib.pyplot as plt

def numerical_derivative(f, x, h=1e-5):
    return (f(x + h) - f(x)) / h
```

## The Mathematics of Gradient Descent

Imagine you are blindfolded on the side of a mountain, and you want to get to the bottom of the valley. How do you do it? You feel the slope of the ground with your feet, and you take a step in the *downward* direction.

In calculus, the derivative tells us the slope. If the slope is positive, the hill goes up to the right. To go down, we must step to the left (subtract). If the slope is negative, the hill goes down to the right. To go down, we must step to the right (add).

We can write this mathematically as:


$$x_{n+1} = x_n - \alpha f'(x_n)$$

Here, $\alpha$ (alpha) is called the **learning rate**. It determines how big of a step you take.
