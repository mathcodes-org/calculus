# Unit 1: Kinematics 

Isaac Newton didn't invent calculus just to draw tangent lines; he invented it to understand the motion of the planets. **Kinematics** is the branch of physics that deals with the motion of objects, and it is entirely built on derivatives.

Here is the golden rule of kinematics:

* If $s(t)$ is the **position** of an object over time...
* The first derivative $s'(t)$ is the **velocity** (how fast position is changing).
* The second derivative $s''(t)$ is the **acceleration** (how fast velocity is changing).

Let's load our tools and simulate some physics!


```python
import numpy as np
import matplotlib.pyplot as plt

def numerical_derivative(f, t, h=1e-5):
    return (f(t + h) - f(t)) / h

# You will also need your second derivative function!
def second_derivative(f, t, h=1e-5):
    return (f(t + h) - 2*f(t) + f(t - h)) / (h**2)
```
