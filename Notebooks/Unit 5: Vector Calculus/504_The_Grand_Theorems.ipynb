We have reached the pinnacle of Vector Calculus.

Up to this point, we calculated **Line Integrals** (walking around a boundary) and **Double Integrals** (summing up an area) as two completely separate concepts. But a mathematician named George Green discovered they are actually two sides of the exact same coin.

**Green's Theorem** (which is the 2D version of Stokes' Theorem) states that if you want to know how much fluid is circulating around the outer boundary of a hurricane, you can just sum up all the microscopic rotation (the Curl) inside the hurricane!


$$\oint_C \vec{F} \cdot d\vec{r} = \iint_R (\nabla \times \vec{F}) dA$$

Let's load our tools and prove this computationally.


```python
import numpy as np
import matplotlib.pyplot as plt

def partial_x(f_matrix, dx):
    return np.gradient(f_matrix, dx, axis=1)

def partial_y(f_matrix, dy):
    return np.gradient(f_matrix, dy, axis=0)
```

### Challenge 3: Faraday's Law of Induction

Maxwell's third equation is Faraday's Law:


$$\nabla \times \vec{E} = -\frac{\partial \vec{B}}{\partial t}$$

In plain English: **If you change a Magnetic Field ($\vec{B}$) over time, it creates a swirling (Curling) Electric Field ($\vec{E}$).** This is how electric generators and wireless chargers work!

**Your Task:**
Imagine a magnetic field pointing straight up out of your screen, and it is rapidly decreasing over time ($\frac{\partial \vec{B}}{\partial t} = -5$). According to Faraday's Law, the Curl of the resulting Electric field must be exactly **+5**.

Because Curl = $5$, the Electric Field is spinning counter-clockwise. Let's define that Electric Field and plot it to see the electricity we just generated!


```python
# If the Curl must be +5, here is the Electric Field that perfectly matches it:
Ex = -2.5 * Y
Ey = 2.5 * X

# Let's plot the electricity generated by our changing magnet!
plt.figure(figsize=(8, 8))
plt.streamplot(X, Y, Ex, Ey, color=np.sqrt(Ex**2 + Ey**2), cmap='plasma')
plt.title("Faraday's Law: Induced Curling Electric Field")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
