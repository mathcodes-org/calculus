# Unit 5: Vector Fields and Fluid Flow

Welcome to Unit 5. Up to this point, our functions have mostly been **Scalar Fields**. A scalar is just a single number. For example, $f(x, y) = x^2 + y^2$ takes a coordinate and gives you back a single height.

But what if we want to model the wind in a hurricane, or the current in a river? At every single coordinate, the wind doesn't have a "height"—it has a **speed** and a **direction**. It is a vector!

A **Vector Field** takes a coordinate $(x, y)$ and outputs a vector $\langle u, v \rangle$, where $u$ is the force in the X-direction and $v$ is the force in the Y-direction.


$$\vec{F}(x, y) = \langle u(x, y), v(x, y) \rangle$$

Let's load our libraries and learn how to draw the wind.
