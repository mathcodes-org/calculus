# Unit 5, Notebook 3: Divergence and Curl

If you drop a microscopic sensor into a flowing river, it will detect two fundamental properties about the water immediately surrounding it:

1. Is the water expanding outward from this point, or compressing inward? This is called **Divergence**.
2. Is the water spinning around this point like a whirlpool? This is called **Curl**.

These two measurements completely define how physical systems behave. In electromagnetism, Divergence describes electric charges (electrons and protons), while Curl describes magnetic fields!

Let's load our libraries and bring back our partial derivative tools from Unit 4 so we can analyze some vector fields.


```python
import numpy as np
import matplotlib.pyplot as plt

# Our trusty 3D derivative tools
def partial_x(f_matrix, dx):
    # Using a fast numpy gradient for matrix derivatives!
    # np.gradient returns the derivative along the axes. For a 2D matrix, axis=1 is the X-axis.
    return np.gradient(f_matrix, dx, axis=1)

def partial_y(f_matrix, dy):
    # axis=0 is the Y-axis
    return np.gradient(f_matrix, dy, axis=0)
```


## 1. Divergence (Sources and Sinks)

The **Divergence** of a vector field $\vec{F} = \langle U, V \rangle$ is calculated by taking the $x$-derivative of the $U$ component, and adding it to the $y$-derivative of the $V$ component.


$$\nabla \cdot \vec{F} = \frac{\partial U}{\partial x} + \frac{\partial V}{\partial y}$$

* If Divergence is **Positive**: The point is a **Source** (fluid is expanding outward, or an explosion occurred).
* If Divergence is **Negative**: The point is a **Sink** (fluid is compressing inward, like a black hole or a drain).
* If Divergence is **Zero**: The fluid is incompressible (like liquid water flowing cleanly through a pipe).

## 2. Curl (Microscopic Rotation)

The **Curl** of a 2D vector field measures how much the fluid wants to spin a microscopic paddlewheel placed in the water.

Notice how the formula looks like a cross product! We take the $x$-derivative of the $V$ component, and subtract the $y$-derivative of the $U$ component:


$$\nabla \times \vec{F} = \frac{\partial V}{\partial x} - \frac{\partial U}{\partial y}$$

* If Curl is **Positive**: The fluid spins counter-clockwise.
* If Curl is **Negative**: The fluid spins clockwise.
* If Curl is **Zero**: The field is irrotational.
