# Unit 2: Convergence and Accumulation

In our last notebook, we calculated the area under a curve using rectangles and trapezoids. But calculus is not just about making good estimates; it is about finding exact answers by taking things to infinity.

Today, we will prove numerically that as the number of shapes approaches infinity, our error drops to zero. Then, we will take a massive conceptual leap: we will view the integral not as a single number, but as a brand-new function.

Let's load our libraries and redefine the `trapezoidal_rule` function we built previously.


```python
import numpy as np
import matplotlib.pyplot as plt

def trapezoidal_rule(f, a, b, n):
    dx = (b - a) / n
    x_edges = np.linspace(a, b, n + 1)
    y_edges = f(x_edges)
    
    left_heights = y_edges[:-1]
    right_heights = y_edges[1:]
    
    areas = ((left_heights + right_heights) / 2) * dx
    return np.sum(areas)
```


## 2. The Accumulation Function

Up until now, we have calculated definite integrals between two fixed numbers, **a** and **b**.

But what if we lock the starting point at **a**, and let the ending point be a variable, **x**? As **x** moves to the right, we accumulate more and more area. This creates a brand-new function, called the **Accumulation Function**:

$$F(x) = \int_0^x f(t) dt$$

Instead of finding the area one time, let's write code to find the cumulative area at hundreds of different points simultaneously.


```python
def f_poly(x):
    return x**2

# Let's look at x-values from 0 to 4
x_vals = np.linspace(0, 4, 100)

# We can calculate the area of hundreds of tiny rectangles using vectorization
# We use a massive number of points so our dx is tiny
x_fine = np.linspace(0, 4, 10000)
dx = x_fine[1] - x_fine[0]

# Calculate the heights at all these points
heights = f_poly(x_fine)

# Find the area of every individual microscopic rectangle
tiny_areas = heights * dx

# Use numpy's cumulative sum feature to add them up as we go!
# This is numerical integration in a single line!
accumulated_areas = np.cumsum(tiny_areas)

print("First 5 tiny areas:", tiny_areas[:5])
print("First 5 accumulated areas:", accumulated_areas[:5])
```
