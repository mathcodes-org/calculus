
# Unit 2: The Fundamental Theorem of Calculus (FTC)

We have spent weeks learning two very different things:

1. **Differentiation:** Finding the instantaneous slope of a curve.
2. **Integration:** Finding the total area under a curve.

To a beginner, these look like completely unrelated branches of math. But the **Fundamental Theorem of Calculus (FTC)** reveals a stunning truth: they are perfect mathematical inverses. Integration is just differentiation performed in reverse.

Today, we will use our code to prove it.


```python
import numpy as np
import matplotlib.pyplot as plt

# Our tool for finding slopes
def numerical_derivative(f_data, x_data):
    """
    Takes arrays of x and y data and returns the slope between each point.
    """
    dy = np.diff(f_data)
    dx = np.diff(x_data)
    return dy / dx

# Our tool for finding accumulation
def numerical_integration(f_func, x_fine):
    """
    Returns the accumulation function (F) for a given function across x_fine.
    """
    dx = x_fine[1] - x_fine[0]
    heights = f_func(x_fine)
    return np.cumsum(heights * dx)

```

## 1. FTC Part 1: The Derivative of an Integral

The first part of the theorem states that if you take a function, integrate it to find the area, and then take the derivative of that area, you end up exactly where you started.

Mathematically: $\frac{d}{dx} \left[ \int_a^x f(t) dt \right] = f(x)$

## 2. FTC Part 2: The Evaluation Theorem

The second part of the theorem gives us a shortcut to find the exact area between $a$ and $b$ without using any rectangles. It says if you know the **antiderivative** ($F$), you can just subtract:


$$\int_a^b f(x)dx = F(b) - F(a)$$

## Summary: Why this matters for Code

Because of the FTC, we have two ways to solve problems:

* If we have a **messy, real-world data set** (like a sensor on a car), we use **numerical integration** (Riemann sums) because we don't have a formula.
* If we have a **clean mathematical equation**, we use the **FTC** to jump straight to the answer.
