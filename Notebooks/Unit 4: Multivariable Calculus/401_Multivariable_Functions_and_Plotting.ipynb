## 1. The Magic of Meshgrid

If we want to plot $f(x) = x^2$, we just make a list of $x$-values (like a ruler) and calculate the heights.

If we want to plot $f(x, y) = x^2 + y^2$, a single ruler isn't enough. We need a floor made of tiles. We need every possible combination of $x$ and $y$ coordinates.

In Python, we create this floor using `np.meshgrid`. It takes two 1D arrays (the $x$-axis and $y$-axis) and weaves them together to create a massive 2D grid of coordinates.


```python
# A tiny x-axis and y-axis
x_tiny = np.array([-1, 0, 1])
y_tiny = np.array([-1, 0, 1])

# Weave them into a grid!
X_grid, Y_grid = np.meshgrid(x_tiny, y_tiny)

print("The X coordinates for our floor:")
print(X_grid)
print("\nThe Y coordinates for our floor:")
print(Y_grid)
```
