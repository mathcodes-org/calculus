# Unit 4: Double Integrals and Volume

In Unit 2, we learned that integration is simply the accumulation of area. We found the area under a 2D curve by slicing it into rectangles and summing them up:


$$Area \approx \sum f(x) \Delta x$$

Now that we are in 3D, we are no longer finding the area under a curve. We are finding the **volume** under a surface. To do this, we slice the volume into 3D rectangular prisms (like tall buildings). The volume of one prism is its height $f(x, y)$ multiplied by its base area ($\Delta x \cdot \Delta y$).

To find the total volume, we must sum up the prisms along the X-axis, and then sum *those* rows along the Y-axis. This is called a **Double Integral**:


$$Volume = \iint_R f(x, y) dx dy \approx \sum \sum f(x, y) \Delta x \Delta y$$
