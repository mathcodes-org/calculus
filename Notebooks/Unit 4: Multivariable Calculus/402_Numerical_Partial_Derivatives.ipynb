# Unit 4: Numerical Partial Derivatives

In Unit 1, we learned that the derivative tells us the slope of a curve. But on a 3D surface, "slope" is ambiguous. If you are standing on the side of a mountain, the slope depends entirely on which direction you are facing!

To handle this, we use **Partial Derivatives**. We lock one variable in place (freezing it as a constant) and take the standard 1D derivative of the other variable.

* **$\frac{\partial f}{\partial x}$** (Partial with respect to $x$): We freeze $y$. This tells us the slope if we step purely East/West.
* **$\frac{\partial f}{\partial y}$** (Partial with respect to $y$): We freeze $x$. This tells us the slope if we step purely North/South.

Let's load our libraries and see how simple this is to code.

## 2. The Gradient Vector

By calculating both partial derivatives, we get a slope for the X-axis and a slope for the Y-axis. If we combine these two numbers into a vector, we get the **Gradient**, written as $\nabla f$ (pronounced "del f").

$$\nabla f(x, y) = \left\langle \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right\rangle$$

The Gradient is arguably the most important concept in multivariable calculus. **It is a vector that points in the direction of the steepest uphill slope.** If you want to climb the mountain as fast as possible, you walk in the direction of the gradient.

Let's calculate the gradient for an entire grid of points.


```python
# A more interesting function: A saddle point (looks like a Pringles chip)
def saddle(x, y):
    return x**2 - y**2

# Create our floor (a dense meshgrid)
x_vals = np.linspace(-3, 3, 50)
y_vals = np.linspace(-3, 3, 50)
X, Y = np.meshgrid(x_vals, y_vals)

# Because we used numpy, we can pass entire matrices into our partial derivative functions!
# u represents the slope in the x-direction for every single point
u = partial_x(saddle, X, Y)

# v represents the slope in the y-direction for every single point
v = partial_y(saddle, X, Y)

print("Shape of our X gradient matrix:", u.shape)
```
