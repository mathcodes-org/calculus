# Unit 4: 3D Optimization (Gradient Descent)

In Unit 1, we used Gradient Descent to find the lowest point of a 2D curve. But real-world problems are rarely 2D.

If you are training an Artificial Intelligence, your "Cost Function" (how many errors the AI is making) might depend on millions of variables. To minimize those errors, you have to find the lowest valley in a multi-dimensional universe.

In our last notebook, we learned that the **Gradient Vector** ($\nabla f$) always points in the direction of the steepest uphill slope. Therefore, if we want to find the lowest valley, we just take our current coordinate and take a step in the exact *opposite* direction of the gradient!

Let's load our libraries and redefine our numerical partial derivatives.


````python
import numpy as np
import matplotlib.pyplot as plt

def partial_x(f, x, y, h=1e-5):
    return (f(x + h, y) - f(x, y)) / h

def partial_y(f, x, y, h=1e-5):
    return (f(x, y + h) - f(x, y)) / h

```
````


## The Mathematics of 3D Descent

In 2D, our update rule was:


$$x_{n+1} = x_n - \alpha f'(x_n)$$

In 3D, we are dealing with points $(x, y)$, which we can write as a vector $\vec{p}$. Our update rule is exactly the same, but we subtract the Gradient Vector:


$$\vec{p}_{n+1} = \vec{p}_n - \alpha \nabla f(\vec{p}_n)$$

This means we update both coordinates simultaneously:

* $x_{new} = x_{old} - \alpha \frac{\partial f}{\partial x}$
* $y_{new} = y_{old} - \alpha \frac{\partial f}{\partial y}$
