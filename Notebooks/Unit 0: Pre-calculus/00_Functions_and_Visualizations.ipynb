# Unit 0, Notebook 1: Functions and Visualizations

Welcome to Calculus! In this course, we aren't just going to memorize formulas; we are going to build the machinery of calculus from scratch using code.

Before we can calculate a derivative or find an integral, we need to know how to speak to the computer. In mathematics, a function is a rule that takes an input, does something to it, and gives an output. In Python, we have tools that do the exact same thing.

First, let's import the two most important libraries for our journey: `numpy` (for fast math) and `matplotlib.pyplot` (for graphing).


```python
import numpy as np
import matplotlib.pyplot as plt
```

## 1. The Math-to-Code Translation

Let's say we have the mathematical function:


$$f(x) = x^2 + 2x - 3$$

In Python, there are two primary ways to write this. The first is using a standard `def` block. This is great for complex functions that require multiple steps.

```python
def f(x):
    return x**2 + 2*x - 3

print("Using def: f(4) =", f(4))
```

The second way is using a `lambda` function. This is a one-liner, perfect for the quick, simple mathematical functions we will use heavily in calculus.

```python
g = lambda x: x**2 + 2*x - 3

print("Using lambda: g(4) =", g(4))
```

## 2. Vectorization: Thinking in Arrays

In calculus, we rarely care about a single point. We care about how a function behaves over an entire domain.

If we wanted to calculate $f(x)$ for numbers 1 through 5, you might be tempted to use a `for` loop. But in Python, loops are slow. Instead, we use `numpy` to generate an array of inputs, and we pass the *entire array* through our function at once. This is called **vectorization**.


```python
# Create an array of x-values from -5 to 5, with 100 points in between
x_vals = np.linspace(-5, 5, 100)

# Pass the entire array into our lambda function!
y_vals = g(x_vals)

# Print the first 5 results to see what happened
print("x values:", x_vals[:5])
print("y values:", y_vals[:5])
```

## 3. Plotting the Math

Now that we have our arrays of $x$ and $y$ values, visualizing the function is incredibly easy using `matplotlib`. Seeing the geometry of a function is crucial for understanding calculus.


```python
plt.figure(figsize=(8, 5))

# Plot the x and y values
plt.plot(x_vals, y_vals, label="f(x) = x^2 + 2x - 3", color='blue')

# Add a horizontal line at y=0 (the x-axis)
plt.axhline(0, color='black', linewidth=1)
# Add a vertical line at x=0 (the y-axis)
plt.axvline(0, color='black', linewidth=1)

# Add grids and labels
plt.grid(True, linestyle='--', alpha=0.7)
plt.title("Graphing our first polynomial")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()

plt.show()
```

## 4. Exploring Domain and Range Visually

Not all functions play nicely with every number. The **domain** is the set of all valid $x$-inputs, and the **range** is the set of all possible $y$-outputs.

Let's look at the square root function:


$$h(x) = \sqrt{x - 4}$$

We know from algebra that we cannot take the square root of a negative number (without getting into complex numbers). So, $x$ must be greater than or equal to 4. Let's see what happens if we graph it and zoom in using `plt.xlim()`.

*(Note: `numpy` will actually throw a warning if we try to give it invalid numbers, so we generate an array that respects the domain!)*

```python
# Create a domain strictly where x >= 4
x_sqrt = np.linspace(4, 10, 100)

# Define the function using numpy's square root
h = lambda x: np.sqrt(x - 4)
y_sqrt = h(x_sqrt)

plt.figure(figsize=(8, 5))
plt.plot(x_sqrt, y_sqrt, label="h(x) = sqrt(x - 4)", color='red')

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

# Force the graph to show x from 0 to 10 to visually highlight the restricted domain
plt.xlim(0, 10)
plt.ylim(-1, 3)

plt.grid(True, linestyle='--', alpha=0.7)
plt.title("Visualizing Restricted Domain")
plt.xlabel("x")
plt.ylabel("h(x)")
plt.legend()

plt.show()
```

Look closely at the graph above. The function literally does not exist to the left of $x = 4$. By graphing functions in Python, we can instantly verify our algebraic intuition about domains and ranges.
