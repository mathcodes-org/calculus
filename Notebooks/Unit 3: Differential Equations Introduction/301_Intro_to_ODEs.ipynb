## 1. Euler's Method (The Bridge Between Units 1 and 2)

If we know a starting point (our initial condition, $y_0$) and we know the equation for the slope (our ODE, $\frac{dy}{dt}$), how do we find the next point?

We use a tangent line! We calculate the slope at our current point, take a tiny step forward in time ($\Delta t$), and calculate our new position. If we do this in a loop, we trace out the mystery function.

This is called **Euler's Method**, and the formula is:


$$y_{n+1} = y_n + f(t_n, y_n) \cdot \Delta t$$

*Notice the connection? It looks exactly like our accumulation formula from numerical integration, but here, the height of our step changes based on our current $y$-value!*

## The Next Frontier: Partial Differential Equations (PDEs)

You have successfully used Python to build calculus from the ground up. You evaluated derivatives, computed integrals, and solved your first ordinary differential equation.

But ODEs only track changes across a **single variable** (usually time). What happens when things change across time *and* space simultaneously?

Welcome to **Partial Differential Equations (PDEs)**.

PDEs run the modern world. They are used to simulate weather systems, the aerodynamics of a Formula 1 car, fluid dynamics in video games, and quantum mechanics. Instead of a single derivative, they use partial derivatives.

Let's look at one of the most famous PDEs, the **2D Heat Equation**:


$$\frac{\partial u}{\partial t} = \alpha \left( \frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} \right)$$

This equation says that the change in temperature over time ($\frac{\partial u}{\partial t}$) depends on how the temperature curves across the X-axis and Y-axis.

To solve it, we apply the exact same concepts you learned in Unit 1 (the difference quotient) and Unit 3 (Euler's method), but expanded into a massive 2D grid. Run the cell below to simulate a spike of heat dissipating across a metal plate over time.

Calculus gives you the rules of the universe; code gives you the power to run them. See you in the next course!


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # Required for 3D plotting

# 1. Setup the spatial grid (a 50x50 metal plate)
N = 50
x = np.linspace(-1, 1, N)
y = np.linspace(-1, 1, N)
X, Y = np.meshgrid(x, y)

# 2. Initial Condition: A massive spike of heat in the center (Gaussian curve)
temperature = np.exp(-15 * (X**2 + Y**2))

# 3. Physics Parameters
alpha = 0.05  # Thermal diffusivity (how fast the metal conducts heat)
dt = 0.01     # Time step
dx = x[1] - x[0]

# 4. Finite Difference Simulation (The PDE Solver)
# We will simulate the heat dissipating for 150 time steps
for _ in range(150):
    # Calculate the second spatial derivatives (concavity) using our Unit 1 concepts!
    d2u_dx2 = (temperature[2:, 1:-1] - 2*temperature[1:-1, 1:-1] + temperature[:-2, 1:-1]) / dx**2
    d2u_dy2 = (temperature[1:-1, 2:] - 2*temperature[1:-1, 1:-1] + temperature[1:-1, :-2]) / dx**2
    
    # Update the temperature of the inner grid points over time
    temperature[1:-1, 1:-1] += alpha * dt * (d2u_dx2 + d2u_dy2)

# 5. The Visual Reveal
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the resulting temperature as a 3D surface
surf = ax.plot_surface(X, Y, temperature, cmap='magma', edgecolor='none', alpha=0.9)

# Formatting for maximum "wow" factor
ax.set_title("2D Heat Equation Simulation", fontsize=16, pad=20)
ax.set_xlabel("X Space", labelpad=10)
ax.set_ylabel("Y Space", labelpad=10)
ax.set_zlabel("Temperature", labelpad=10)

# Remove the gray background panes for a cleaner, modern look
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False
ax.grid(False)

# Add a color bar map
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, label='Heat Intensity')

plt.show()
```
